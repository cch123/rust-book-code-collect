```python
from matplotlib import pyplot as plt
```

```python
size = [1000, 10000, 100000, 1000000, 10000000]
skip_list = [311, 438, 1190, 2609, 3334]
linked_list = [825, 17574, 428259, 5440420, 45157562]
```

```python
plt.figure(figsize=(16,20))
plt.subplot(211)
plt.title("Skip List find()")
plt.ylabel("ns")
plt.xlabel("List size")
plt.plot(size, skip_list, color="blue")
plt.subplot(212)
plt.title("Linked List find()")
plt.ylabel("ns")
plt.xlabel("List size")
plt.plot(size, linked_list, color="red")
```
